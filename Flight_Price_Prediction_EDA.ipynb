```python
# Feature Engineering on important fetures 
# 1. Date_of_Journey, convert into date time format, we are trying to create a derived feature
final_df['Date'] = final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month'] = final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year'] = final_df['Date_of_Journey'].str.split('/').str[2]
```

```python
final_df.info()
```

```python
final_df.head()
```

```python
final_df.info()
```

```python
final_df['Date'] = final_df['Date'].astype(int)
final_df['Month'] = final_df['Month'].astype(int)
final_df['Year'] = final_df['Year'].astype(int)
final_df.info()
```

```python
final_df.drop('Date_of_Journey', axis = 1, inplace = True)
```

```python
final_df.head(1)
```

```python
# focusing on arrival time and departure time, try focusing on similar features to convert 
final_df['Arrival_Time'].str.split(' ').str[0] # since we already have the dates 
```

```python
final_df['Arrival_Time'] = final_df['Arrival_Time'].str.split(' ').str[0]
```

```python
final_df.isnull().sum()
```

```python
final_df.info()
```

```python
# we need to take out the hour and minutes
final_df['Arrival_Hour'] = final_df['Arrival_Time'].str.split(':').str[0]
final_df['Arrival_Minute'] = final_df['Arrival_Time'].str.split(':').str[1]
final_df.head()
```

```python
final_df['Arrival_Hour'] = final_df['Arrival_Hour'].astype(int)
final_df['Arrival_Minute'] = final_df['Arrival_Minute'].astype(int)
final_df.info()
```

```python
final_df.drop('Arrival_Time', axis =1, inplace = True)
final_df.head(1)
```

```python
final_df['Dep_Hour'] = final_df['Dep_Time'].str.split(':').str[0]
final_df['Dep_Minute'] = final_df['Dep_Time'].str.split(':').str[1]
final_df.head(5)
```

```python
final_df['Dep_Hour'] = final_df['Dep_Hour'].astype(int)
final_df['Dep_Minute'] = final_df['Dep_Minute'].astype(int)
final_df.info()
```

```python
final_df.drop('Dep_Time', axis = 1, inplace = True)
final_df.head(1)
```

```python
final_df['Route'].unique()
```

```python
# The route data has a very high cardinality, we ca see that we have 
# source and destination and no of stops, cause price would be affected by no of stops
# we can consider the route parameter
final_df['Total_Stops'].unique()
```

```python
final_df[final_df['Total_Stops'].isnull()]
```

```python
# we see that the missing vlaue has a flight to dehli to Cochin we can assume that it has two stops
```

```python
# so since stops categorical and ordinal, we can use ordinal mapping to convert it 
final_df['Total_Stops'] = final_df['Total_Stops'].map({'non-stop':0,'1 stop':1, '2 stops':2, '3 stops':3, '4 stops':4, 'nan':2})
final_df.head()
```

```python
final_df.info()
```

```python
final_df.drop('Route', axis = 1, inplace = True)
```

```python
final_df.head()
```

```python
final_df['Total_Stops'].isnull().sum()
```

```python
final_df['Additional_Info'].unique()
```

```python
# Since Additional_Info has categorical data not ordinal, we can use one hot encoding but take the decision based on 
# the cardinatlity of this feature 
info_add = pd.get_dummies(final_df['Additional_Info'], drop_first = True, dtype = int)
```

```python
final_df = pd.concat([final_df, info_add], axis =1)
```

```python
final_df.drop('Additional_Info', axis=1, inplace=True)
final_df.head()
```

```python
final_df.drop(6474, axis = 0, inplace = True)
final_df.drop(2660, axis = 0, inplace = True) # both of these do not make sense cause you cannot reach Mumbai to HYD in 5 minutes 
```

```python
# Now considering duration, convert all of into minutes 
final_df['Duration_Hour'] = final_df['Duration'].str.split(' ').str[0].str.split('h').str[0]
```

```python
final_df[final_df['Duration']=='5m'] 
```

```python
final_df['Duration_Hour'] = final_df['Duration_Hour'].astype(int)
```

```python
final_df['Duration_Hour'] = final_df['Duration_Hour'] * 60
```

```python
# converting the minutes
final_df['Duration_Minute'] = final_df['Duration'].str.split(' ').str[1].str.split('m').str[0]
```

```python
final_df[final_df['Duration_Minute'].isnull()]
```
